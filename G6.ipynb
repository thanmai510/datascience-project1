```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(a)
print(myvar)
print('----------------')
#labels
print(myvar[0])
print('----------------')
#assigning seperate labels to the vales
myvar2=pd.Series(a,index=["x","y","z"])
print(myvar2)
print(myvar2["y"])
print('------------')
#series generated from dictionrary
calories={"day1":420,"day2":380,"day3":390}
myvar=pd.Series(calories)
print(myvar)
print('----------')
#series generated from scalar values
series=pd.Series(5,index=[0,1,2,3])
print(series)
print('----------------------')
```

```python
import pandas as pd
#integer series
int_series=pd.Series([1,2,3,4])
print(int_series)
#float series
float_series=pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
#string series
string_series=pd.Series(['a','b','c','d'])
print(string_series)
#label-based indexing
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series['a'])
print(series[['a','c']])
#boolean indexing
data=[10,20,30,40,50]
series=pd.Series(data)
print(series[series>25])
```

```python
import pandas as pd
#creating a data frames
data={
    'name':['alice','bob','charlie'],
    'age':[20,30,35],
    'city':['newyork','losangles','chicago']
}
df=pd.DataFrame(data)
print(df)
print('--------------')
#selecting column
print(df.name)
print(df['name'])
print(df[['name','city']])
print('-----------------')
#select a single row by label
print(df.loc[0])
print(df.iloc[0])
#slect multiple rows
print(df.iloc[0:2])
print('-------------------')
#select rows where age is greater than 25
print(df[df['age']>25])
print('-----------------')
print(df.loc[(df.name=='bob')&(df.city=='los angeles')])
```

```python
import pandas as pd
#grouping by a single column
data={
    'name':['alice','bob','charlie','alice','bob'],
    'age':[25,30,35,25,30],
    'city':['newyork','san fransisco','los angles','newyork','san fransisco'],
    'salary':[70000,80000,120000,70000,80000]

}
df=pd.DataFrame(data)
grouped=df.groupby('name')
print(grouped.groups)
#calculate the sum of salaries for each group
salary_sum=grouped['salary'].sum()
print(salary_sum)
age_mean=grouped['age'].sum()
print(age_mean)
#groupby multiple coulumns
grouped_multiple=df.groupby(['name','city'])
salary_sum_multiple=grouped_multiple['salary'].sum()
print(salary_sum_multiple)
```
